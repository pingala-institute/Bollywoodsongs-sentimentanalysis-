```python
import pandas as pd
from textblob import TextBlob

# Step 1: Load the CSV data into a pandas DataFrame
df = pd.read_csv(r"C:\Users\rajsi\OneDrive\Desktop\git_unsw\Bollywoodsongs-sentimentanalysis-\data\cleaned_and_translated_file_and_cosine.csv")

# Step 2: Preprocessing (assuming 'Lyrics' and 'translated_Lyrics' columns exist)
# Assuming your CSV has columns 'Lyrics' for Hindi lyrics and 'translated_Lyrics' for English translations

# Step 3: Sentiment Analysis using TextBlob (Hindi language support)
def analyze_sentiment(text):
    analysis = TextBlob(text)
    # Return sentiment polarity (ranges from -1 to 1 where -1 is very negative and 1 is very positive)
    return analysis.sentiment.polarity

# Apply sentiment analysis to each row in 'Lyrics' and 'translated_Lyrics' columns
df['Hindi_Sentiment'] = df['Lyrics'].apply(analyze_sentiment)
df['English_Sentiment'] = df['translated_Lyrics'].apply(analyze_sentiment)

# Step 4: Optionally, you can visualize the sentiment scores
import matplotlib.pyplot as plt

plt.hist(df['Hindi_Sentiment'], bins=20, alpha=0.7, label='Hindi Sentiment')
plt.hist(df['English_Sentiment'], bins=20, alpha=0.7, label='English Sentiment')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.title('Sentiment Analysis of Hindi Songs')
plt.legend()
plt.show()
```

```python
# Function to back-translate text
from deep_translator import GoogleTranslator
def hindi_translate_text(text, source_language='auto', target_language='hi'):
    try:
        translator = GoogleTranslator(source=source_language, target=target_language)
        translation = translator.translate(text)
        return translation
    except Exception as e:
        print(f"Error back-translating text: {e}")
        return text

# Back-translate the 'Lyrics' column
df['hindi_Translated_Lyrics'] = df['Lyrics'].apply(lambda x: hindi_translate_text(x) if pd.notna(x) else x)

# Display the first few rows of the dataframe with back-translated lyrics
print(df[['Lyrics', 'hindi_Translated_Lyrics']].head())

# Save the dataframe with back-translated lyrics to a new CSV file
# df.to_csv('data/translated_and_back_translated_file.csv', index=False)

print(f'Back-translated lyrics saved to data/translated_and_back_translated_file.csv')
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Create a SentimentIntensityAnalyzer object
analyzer = SentimentIntensityAnalyzer()

# Define a function to get sentiment scores using VADER
def analyze_sentiment_vader(text):
    sentiment_scores = analyzer.polarity_scores(text)
    return sentiment_scores['compound']  # Using compound score for overall sentiment

# Apply sentiment analysis to Hindi lyrics
df['English_Sentiment_VADER'] = df['translated_Lyrics'].apply(analyze_sentiment_vader)
```

```python
plt.hist(df['Hindi_Sentiment'], bins=20, alpha=0.7, label='Hindi Sentiment')
plt.hist(df['English_Sentiment'], bins=20, alpha=0.7, label='English Sentiment')
plt.hist(df['Hindi_Sentiment_VADER'], bins=20, alpha=0.7, label='Hindi Sentiment VADER')
plt.hist(df['English_Sentiment_VADER'], bins=20, alpha=0.7, label='english Sentiment VADER')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.title('Sentiment Analysis of Hindi Songs')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import mean_squared_error, mean_absolute_error
from scipy.stats import pearsonr

# Generate confusion matrix
mse = mean_squared_error(df['English_Sentiment'], df['English_Sentiment_VADER'])

print(f"Mean Squared Error (MSE): {mse:.2f}")

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(df['English_Sentiment'], df['English_Sentiment_VADER'])

print(f"Mean Absolute Error (MAE): {mae:.2f}")

# Calculate Pearson correlation coefficient
pearson_corr, _ = pearsonr(df['English_Sentiment'], df['English_Sentiment_VADER'])
```

```python
def convert_to_category(score):
    if score < 0:
        return 'negative'
    elif score > 0:
        return 'positive'
    else:
        return 'neutral'

# Apply conversion function to create categorical columns
df['TextBlob_Sentiment_Category'] = df['English_Sentiment'].apply(convert_to_category)
df['VADER_Sentiment_Category'] = df['English_Sentiment_VADER'].apply(convert_to_category)
```
