```python
import pandas as pd
from transformers import pipeline, AutoTokenizer, TFAutoModelForSequenceClassification

# Read the CSV file
df = pd.read_csv(r'C:\Users\rajsi\OneDrive\Desktop\git_unsw\Bollywoodsongs-sentimentanalysis-\data\cleaned_and_translated_file.csv')

# Specify the model name explicitly
model_name = "distilbert-base-uncased-finetuned-sst-2-english"  # Change this if you have a different model

# Load the tokenizer and TensorFlow model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = TFAutoModelForSequenceClassification.from_pretrained(model_name)

# Create a sentiment analysis pipeline using TensorFlow
sentiment_pipeline = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer, framework='tf')

# Define a function to get the sentiment, handling long texts by truncating
def get_sentiment(text):
    # Tokenize and truncate if text length exceeds model's maximum length (512 tokens)
    inputs = tokenizer(text, return_tensors="tf", truncation=True, max_length=512)
    result = model(inputs)[0]
    label = model.config.id2label[result.numpy().argmax()]
    score = result.numpy().max()
    return label, score

# Apply the function to the 'Lyrics' column
df['sentiment'], df['sentiment_score'] = zip(*df['Lyrics'].apply(get_sentiment))

# Save the modified DataFrame back to a CSV file
df.to_csv(r'C:\Users\rajsi\OneDrive\Desktop\git_unsw\Bollywoodsongs-sentimentanalysis-\data\lyrics_with_sentiment.csv', index=False)

# Display the first few rows of the modified DataFrame
print(df.head(10))
```

```python
import matplotlib.pyplot as plt

# Example assuming df is your DataFrame with sentiment scores
plt.figure(figsize=(8, 6))
plt.hist(df['sentiment_score'], bins=30, edgecolor='black')
plt.title('Distribution of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.groupby('year')['sentiment_score'].mean().plot(marker='o')
plt.title('Average Sentiment Score Over Time')
plt.xlabel('year')
plt.ylabel('Average Sentiment Score')
plt.grid(True)
plt.xticks(df['year'].unique(), rotation=45)  # Adjust rotation angle here
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sentiment_counts = df['sentiment'].value_counts()
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Sentiment Labels')
plt.axis('equal')
plt.show()
```

```python
import seaborn as sns

# Assuming 'Decade' is a new column derived from 'Year'
df['Decade'] = (df['year'] // 10) * 10

plt.figure(figsize=(10, 6))
sns.boxplot(x='Decade', y='sentiment_score', data=df)
plt.title('Distribution of Sentiment Scores by Decade')
plt.xlabel('Decade')
plt.ylabel('Sentiment Score')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Example assuming df is your DataFrame with sentiment analysis results
positive_lyrics = ' '.join(df[df['sentiment'] == 'POSITIVE']['Lyrics'])
negative_lyrics = ' '.join(df[df['sentiment'] == 'NEGATIVE']['Lyrics'])
neutral_lyrics = ' '.join(df[df['sentiment'] == 'NEUTRAL']['Lyrics'])

# Check if the text is not empty before generating word clouds
if positive_lyrics:
    plt.figure(figsize=(8, 6))
    wordcloud_positive = WordCloud(width=800, height=400, background_color='white').generate(positive_lyrics)
    plt.imshow(wordcloud_positive, interpolation='bilinear')
    plt.title('Positive Sentiment Word Cloud')
    plt.axis('off')
    plt.show()
else:
    print("No positive sentiment lyrics to generate word cloud.")

if negative_lyrics:
    plt.figure(figsize=(8, 6))
    wordcloud_negative = WordCloud(width=800, height=400, background_color='white').generate(negative_lyrics)
    plt.imshow(wordcloud_negative, interpolation='bilinear')
    plt.title('Negative Sentiment Word Cloud')
    plt.axis('off')
    plt.show()
else:
    print("No negative sentiment lyrics to generate word cloud.")

if neutral_lyrics:
    plt.figure(figsize=(8, 6))
    wordcloud_neutral = WordCloud(width=800, height=400, background_color='white').generate(neutral_lyrics)
    plt.imshow(wordcloud_neutral, interpolation='bilinear')
    plt.title('Neutral Sentiment Word Cloud')
    plt.axis('off')
    plt.show()
else:
    print("No neutral sentiment lyrics to generate word cloud.")
```
