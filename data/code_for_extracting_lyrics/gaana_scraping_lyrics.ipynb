```python
import re
from bs4 import BeautifulSoup 
import pandas as pd
import requests
```

```python
#seaching songs whose lyrics are not found through gaana web

def getlyrics(song_name):
    song_name = song_name.lower()
    url="https://gaana.com/song/"+song_name.replace(" ","-").strip()

    try: 
        page=requests.get(url)
        soup =BeautifulSoup(page.content, "html.parser" )
        scraped= soup.find_all('div', class_='data')
        lyrics=scraped[0].get_text().strip()
        return lyrics 
    except Exception as e: 
        return "Lyrics not found"
```

```python
file_path = "D:\Research_unsw\data - Copy.xlsx"
songs_df = pd.read_excel(file_path)
songs_df
```

```python
for index, row in songs_df.iterrows():
    song_title = row['songs']
    print("searching for"+ song_title)
    lyrics = getlyrics(song_title)
    if(lyrics=="Lyrics not found"):
        print("Lyrics not found")
    else:
        songs_df.at[index, 'Lyrics'] = lyrics
        print("done")
```

```python
import lyricsgenius
import re

api_key = 'J7AkpD0XK0JzorAkobj188C2WJMwkfVJCapeEFCfi3wcICT5O1TdwiuqKDeWNb2Y'
genius = lyricsgenius.Genius(api_key)

def clean_lyrics(lyrics):
    lines = lyrics.split('\n')
    cleaned_lyrics = '\n'.join(lines[1:]).strip()
    # Remove bracketed text such as [Verse 1], [Chorus], etc.
    cleaned_lyrics = re.sub(r'\[.*?\]', '', lyrics)
    # Remove any leading/trailing whitespace
    cleaned_lyrics = cleaned_lyrics.strip()
    
    return cleaned_lyrics

def get_song_lyrics(song_title, artist_name):
    try:
        song = genius.search_song(song_title, artist_name)
        if song:
            lyrics = clean_lyrics(song.lyrics)
            return lyrics
        else:
            return "Lyrics not found"
    except Exception as e:
        print(f"Error fetching lyrics for {song_title} by {artist_name}: {e}")
        return "Lyrics not found"
```

```python
def try_remaining_artists(song_title, artists):
    for artist in artists.split(','):
        lyrics = get_song_lyrics(song_title, artist)
        if lyrics != "Lyrics not found":
            return lyrics
    return "Lyrics not found"
```

```python
def remove_first_line(lyrics):
    if lyrics and lyrics != "Lyrics not found":
        # Split the lyrics into lines and join them back, skipping the first line
        lines = lyrics.split('\n')
        cleaned_lyrics = '\n'.join(lines[1:]).strip()
        return cleaned_lyrics
    else:
        return lyrics

# Apply the function to the 'Lyrics' column
songs_df['Lyrics'] = songs_df['Lyrics'].apply(remove_first_line)
```

```python
import numpy as np

def replace_new_lines_with_full_stops(lyrics):
    if isinstance(lyrics, str) and lyrics != "Lyrics not found":
        # Replace new line characters with full stops
        cleaned_lyrics = lyrics.replace('\n', ',').strip()
        return cleaned_lyrics
    else:
        return lyrics

# Apply the function to the 'Lyrics' column
songs_df['Lyrics'] = songs_df['Lyrics'].apply(replace_new_lines_with_full_stops)
```
