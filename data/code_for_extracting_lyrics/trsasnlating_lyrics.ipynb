```python
import pandas as pd
from deep_translator import GoogleTranslator

# Load the CSV file
df = pd.read_csv('data/cleaned_file.csv')

# Function to translate text using deep-translator
def translate_text(text, target_language='en'):
    try:
        translator = GoogleTranslator(source='auto', target=target_language)
        translation = translator.translate(text)
        return translation
    except Exception as e:
        print(f"Error translating text: {e}")
        return text

# Translate the 'lyrics' column
df['Lyrics'] = df['Lyrics'].apply(lambda x: translate_text(x) if pd.notna(x) else x)

# Save the cleaned and translated dataframe to a CSV file in the 'data' folder
# df.to_csv('data/cleaned_and_translated_file.csv', index=False)
```
