```python
import string
```

```python
def process_lyrics(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text
```

```python
df['Lyrics'] = df['Lyrics'].apply(process_lyrics)
```
